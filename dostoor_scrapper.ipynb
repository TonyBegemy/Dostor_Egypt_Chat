```python
bab_count = 6
bab_titles = ["الديباجة", "الباب الأول: الدولة", "الباب الثالث: الحقوق والحريات والواجبات العامة", "الباب الرابع: سيادة القانون", "الباب السادس: الأحكام العامة والانتقالية", "الباب السابع: مجلس الشيوخ"]
# bab_titles_cont_counter = -1
# bab_titles_cont = [" الباب الثاني: المقومات الأساسية للمجتمع", "الباب الخامس: نظام الحكم"]
```

```python
full_text = ""
for i in range(bab_count):
    if i > 0:
        bab_title = bab_titles[i]
        bab_text = right_titles_section.find_all("div", class_="single-section")[i]
    
        #clean the shit
        bab_text = str(bab_text).replace('&lt;', '<').replace('&gt;', '>')
        #the holy cleaning
        bab_text = bab_text.replace('<div>', '').replace('</div>','').replace('<p>','').replace('</p>','').\
        replace('&quot;','').replace('</h2>','').replace('</a>','').replace('<a>','').replace('<br />','').\
        replace('<h2 />','').replace("<div class='article-content'>",'').replace("<div class='single-section active-acc' style='font-size: 16px;'>",'').\
        replace("<div class='article-content'>",'').replace("<div class='header-description'>",'').replace("<h2 class='header-top-bottom-line'>",'')\
        .replace("<div class='row'>",'').replace("<div class='col-md-12 articles-container'>",'').replace("<div class='header-content'>",'')\
        .replace("<p style=text-align: center;>",'').replace("<div class='col-md-12 articles-container'>",'').replace("<div class='header-content'>",'')\
        .replace('<div class="single-section active-acc" style="font-size: 16px;">','').replace("<div class='col-md-12 articles-container'>",'').replace("<div class='article-container'>",'')\
        .replace("<div class='container-fluid main-articles-container'>",'').replace('<a class="btn-primary-1 show-articles" data-text="','').replace("<div class='article-container'>",'')\
        .replace('" style="font-size: 14px;">','').replace("<div class='col-md-12 articles-container'>",'').replace("<div class='header-content'>",'')\
        .replace('<div class="single-section" style="font-size: 16px;">','')
        
        full_text += bab_text.replace("\n",'')
        full_text += "---------"
        print(full_text)
        print("\n\n\n---------\n\n\n")
    
```

```python
array = [1, 3, 4, 6, 7]

# Split the string into parts using regex to identify "مادة" followed by a number
parts = re.split(r'(مادة \(\d+\))', full_text)

# Initialize the dataframe
data = []
current_bab_index = 0
current_bab = array[current_bab_index]

# Process each part
for i in range(1, len(parts), 2):
    madda = re.search(r'\d+', parts[i]).group(0)
    text = parts[i+1].split('مادة', 1)[0].strip() if (i + 1) < len(parts) else ""
    
    data.append({
        'bab': current_bab,
        'madda': madda,
        'Text': text
    })

    if "---------" in text:
        current_bab_index += 1
        if current_bab_index < len(array):
            current_bab = array[current_bab_index]

df = pd.DataFrame(data)
df
```
